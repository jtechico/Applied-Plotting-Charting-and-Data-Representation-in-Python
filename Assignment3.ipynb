In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
dfT =df.T #Transpose the table to get stats by year
dfStats = dfT.describe() #Stats by year
dfStatsT = dfStats.T #Transpose Stats to compute and add y error per year
dfStatsT['yerr'] = 1.96 * dfStatsT['std'] / np.sqrt(dfStatsT['count']) #Compute Y error
dfStats = dfStatsT.T #Transpose Stats back to portray by year
dfStats #Show data
```

```python
xList = list(dfStats) #Year on the X Axis
yList = list(dfStats.loc['mean']) #Means on the Y Axis
errList = list(dfStats.loc['yerr'])

pos = np.arange(len(xList))/1.5 #allows to position the bar width
fig= plt.figure() #Plot the figure

container = plt.bar(pos, yList, align='center', yerr=errList,
                    edgecolor='black', color=['blue', 'white', 'white', 'red'],
                    width = .5, linewidth =1)
_=plt.xticks(pos, xList) #Plot tick marks on the x axis with the pos variable, and year label

bars = container.get_children() #call bars the number of children objects
for b in bars:
    b.set_color((1, 0, 0, .75)) #Set the bars to an initial color RGBA

#Line settings
y=39493 #Start the line at 39493
x1, x2 = plt.xlim() #Set the x variables to the current axes
line, = plt.plot([x1, x2], [y, y], '-', color=(0,0,0,.5), lw=2)

def onclick(event):
    plt.gca().set_title('Distribution comparison: {}'.format(event.ydata))
    line.set_ydata([event.ydata,event.ydata]) #Creates a line at the y value selected
    yPick = event.ydata #the y value selected
    for b, yMean, yErr in zip(bars, yList, errList): #impact each bar
        redVal = 0
        blueVal = 0
        alpha = 1
        if yPick > yMean : #If the y value selected is greater than the mean, shade blue
            blueVal = 1
            alpha = min(1, (yPick-yMean)/yErr) #change the alpha as you get closer to mean
        if yPick < yMean : #If the y value selected is smaller than the mean, shade red
            redVal = 1
            alpha = min(1, (yMean-yPick)/yErr) #change the alpha as you get closer to mean
        b.set_color((redVal, 0, blueVal, alpha)) #Set the color of the bar according to RGBA
        b.set_edgecolor('black') #set the edge color of each bar

plt.gcf().canvas.mpl_connect('button_press_event', onclick) #Pass the button press event into onclick
```
